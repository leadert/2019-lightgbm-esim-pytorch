```python
import gensim
from gensim.models import Word2Vec
import time
import os
import csv

def data_to_sentence():
    if not os.path.exists("lgb_sentences"):
        f = open("lgb_sentences", "w")
        f.close()

    with open("lgb_sentences", "w") as f:
        with open("/home/kesci/temp/train_tail_50M.csv") as csv_file1:                    # 训练集
            csv_reader = csv.reader(csv_file1)
            initial_query_id = 0

            for row in csv_reader:
                query_id = row[0]
                query = row[1].replace("\t", "")
                title = row[3].replace("\t", "")

                if int(query_id) != initial_query_id:
                    f.write(query + "\n")
                    f.write(title + "\n")
                    initial_query_id = int(query_id)

                else:
                    f.write(title + "\n")
        csv_file1.close()

        with open("/home/kesci/input/bytedance/bytedance_contest.final_2.csv") as csv_file2:     # 测试集
            csv_reader = csv.reader(csv_file2)
            initial_query_id = 0

            for row in csv_reader:
                query_id = row[0]
                query = row[1].replace("\t", "")
                title = row[3].replace("\t", "")

                if int(query_id) != initial_query_id:
                    f.write(query + "\n")
                    f.write(title + "\n")
                    initial_query_id = int(query_id)

                else:
                    f.write(title + "\n")
        csv_file2.close()
    f.close()

    print("Writing sentence Successful!")


data_to_sentence()
```

```python
class EpochSaver(gensim.models.callbacks.CallbackAny2Vec):
    '''用于保存模型, 打印损失函数等等'''

    def __init__(self, save_name="lgb_word2vec.model"):
        self.save_path = save_name
        self.epoch = 0
        self.pre_loss = 0
        self.best_loss = 999999999999.9
        self.since = time.time()

    def on_epoch_end(self, model):
        self.epoch += 1
        cum_loss = model.get_latest_training_loss()  # 返回的是从第一个epoch累计的
        epoch_loss = cum_loss - self.pre_loss
        time_taken = time.time() - self.since
        print("Epoch %d, loss: %.2f, time: %dmin %ds" % (self.epoch, epoch_loss, time_taken // 60, time_taken % 60))
        if self.best_loss > epoch_loss:
            self.best_loss = epoch_loss
            print("Better model. Best loss: %.2f" % self.best_loss)
            model.save(self.save_path)
            print("Model %s save done!" % self.save_path)

        self.pre_loss = cum_loss
        self.since = time.time()

 
model = Word2Vec(min_count=1, window=5, size=150, workers=4)
model.build_vocab(corpus_file="lgb_sentences")
model.train(corpus_file="lgb_sentences", total_words=model.corpus_total_words,
            epochs=5, compute_loss=True, report_delay=60 * 10,
            callbacks=[EpochSaver("lgb_word2vec.model")])
```
